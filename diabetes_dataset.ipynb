```python
from sklearn import datasets
diabetes=datasets.load_diabetes()
```

```python
import pandas as pd
df=pd.DataFrame(diabetes.data)
df.head()
```

```python
df.columns=diabetes.feature_names
df.head()
```

```python
df.describe()
```

```python
#creating distribution time graph
import seaborn as sns
import matplotlib.pyplot as plt
for i in df.columns:
  sns.displot(df[i])
  plt.title(i)
  plt.show()
```

```python
for i in df.columns:
  print(i, "Skewness: ", df[i].skew())
  print(i, "Kurtosis: ", df[i].kurt())
```

```python
for i in df.columns:
  print(i,  "Median",df[i].median())
  print(i, "Mean", df[i].mean())
  print(i, "mode", df[i].mode())

```
